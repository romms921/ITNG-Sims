```python
data = pl.read_csv('/Volumes/T7 Shield/Simulation Output/System 1/Sim 1.csv')
data
```

```python
# Reduced data for testing
data = data[:200000]
```

```python
vmin = 0
vmax = 100
norm = plt.Normalize(vmin=vmin, vmax=vmax)
plt.scatter(data['pos_rms'], data['chi2'], c=data['mag_rms'], cmap='viridis', s=1, norm=norm)
plt.xlim(0, 0.01)
plt.colorbar(label='Magnification RMS')
plt.xlabel('Position RMS (arcsec)')
plt.ylabel('Chi-Squared')
plt.title('Position RMS vs Chi-Squared Colored by Magnification RMS')
plt.yscale('log')
```
